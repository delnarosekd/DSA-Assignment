# 1. count plot of outcome

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x='Outcome',data=auto_data)
plt.show()

```

# 2. Effect of age on outcome using boxplot

```python
import matplotlib.pyplot as plt
sns.boxplot(x=auto_data['Outcome'],y=auto_data['Age'],data=auto_data)
plt.title('effect of age on outcome')
plt.show
```

# 3. Chance of getting diabetes: violin plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.violinplot(y=auto_data['DiabetesPedigreeFunction'],x=auto_data['Outcome'],data=auto_data)
plt.show
```

# 4. Relationship with insulin and glucose: SCATTER PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.scatterplot(x=auto_data['Insulin'],y=auto_data['Glucose'],data=auto_data,hue=auto_data['Outcome'])
plt.show
```

# 5. Heatmap 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.heatmap(auto_data.corr())
plt.show
```
